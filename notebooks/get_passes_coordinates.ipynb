# Get passes coordinates

```python
import pandas as pd
import sportslabkit as slk
import numpy as np
```

```python
passes_frame = pd.read_csv('../all_passe.csv', index_col=0)
passes_frame = passes_frame.loc[passes_frame['start_passe'] != passes_frame['end_passe'], :]
passes_frame
```

```python
df_coordinates = slk.load_df('../all_players_2D.csv', df_type='coordinates')
df_coordinates
```

```python
passes_frame = passes_frame.assign(x = 0, y = 0, end_x = 0, end_y = 0)
passes_frame
```

```python
df_coordinates.xs(('BALL', 'BALL'), level=('TeamID', 'PlayerID'), axis=1)
```

```python
ball_coordinates = df_coordinates.xs(('BALL', 'BALL'), level=('TeamID', 'PlayerID'), axis=1)

for index, row in passes_frame.iterrows():
    start_frame = row['start_passe']
    end_frame = row['end_passe']
    
    passes_frame.at[index, 'x'] = ball_coordinates.loc[start_frame].x
    passes_frame.at[index, 'y'] = ball_coordinates.loc[start_frame].y
    
    passes_frame.at[index, 'end_x'] = ball_coordinates.loc[end_frame].x
    passes_frame.at[index, 'end_y'] = ball_coordinates.loc[end_frame].y
```

```python
passes_frame
```

## Get the players who have the ball

d = √[(x₂ - x₁)² + (y₂ - y₁)²]

```python
passes_frame = passes_frame.assign(player_start = np.nan, player_end = np.nan)
passes_frame
```

```python
for index, row in passes_frame.iterrows():
    start = {
        'frame': row["start_passe"],
        'x1': row["x"],
        'y1':row["y"],
        'min_dist': 9999999,
        'player': -1
    }
    end = {
        'frame': row["end_passe"],
        'x1': row["end_x"],
        'y1':row["end_y"],
        'min_dist': 9999999,
        'player': -1
    }
    
    
    for player_index, row in df_coordinates.loc[start['frame'], (slice('0', '1'))].unstack(level=2).iterrows():
        d = ((row['x'] - start['x1'])**2 + (row['y'] - start['y1'])**2) ** (0.5)
        if (d < start['min_dist']):
            start['min_dist'] = d
            start['player'] = player_index
    
    for player_index, row in df_coordinates.loc[end['frame'], (slice('0', '1'))].unstack(level=2).iterrows():
        d = ((row['x'] - end['x1'])**2 + (row['y'] - end['y1'])**2) ** (0.5)
        if (d < end['min_dist']):
            end['min_dist'] = d
            end['player'] = player_index
    
    passes_frame.at[index, 'player_start'] = start['player']
    passes_frame.at[index, 'player_end'] = end['player']
```

```python
passes_frame['team_start'] = passes_frame['player_start'].apply(lambda x: x[0])
passes_frame['team_end'] = passes_frame['player_end'].apply(lambda x: x[0])
```

```python
passes_frame
```

```python
passes_frame["successful"] = passes_frame["team_start"] == passes_frame["team_end"]
```

```python
passes_frame = passes_frame.loc[(passes_frame.end_x != 0) & (passes_frame.end_y) != 0, :]
passes_frame = passes_frame.loc[passes_frame.player_start != passes_frame.player_end, :]
passes_frame
```
