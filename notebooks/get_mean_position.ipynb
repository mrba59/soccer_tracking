# Get mean position for each player

```python
import pandas as pd
import sportslabkit as slk
```

```python
df = slk.load_df('../all_players_2D.csv', df_type='coordinates')
df
```

```python
df.columns
```

```python
positions = df.loc[:, slice('0', '1')].mean().unstack(level=[2]).reset_index()
positions
```

```python
positions["team_player"] = positions.apply(lambda x: (x["TeamID"], x["PlayerID"]), axis=1)
positions["pass_count"] = 0
positions
```

Add number of pass received or initiate

```python
pass_df = pd.read_csv('../dashboard/src/stats/all_passe_with_coordinates.csv', index_col=0)
pass_df
```

```python
for index, row in positions.iterrows():
    nb_pass = pass_df.loc[
        ((pass_df["player_start"] == str(row["team_player"])) | (pass_df["player_end"] == str(row["team_player"]))) 
        & (pass_df.successful == True), 'start_passe'].count()
    positions.at[index, 'pass_count'] = nb_pass
    
positions
```

```python
positions.to_csv('../dashboard/src/stats/mean_position.csv')
```

get all pass between two players

```python
pass_df = pass_df.loc[pass_df.successful == True]
pass_relation = pass_df.groupby(['player_start', 'player_end']).count().loc[:,'start_passe']
pass_relation
```

```python
pass_relation = pass_relation.reset_index()
pass_relation = pass_relation.rename(columns={'start_passe': 'pass_count'})
pass_relation.head(20)
```

Remove the lines in twice (the relation passer/receiver is the same as receiver/passer)

```python
pass_relation["pos_min"] = pass_relation.apply(lambda x: min(x["player_start"], x["player_end"]), axis=1)
pass_relation["pos_max"] = pass_relation.apply(lambda x: max(x["player_start"], x["player_end"]), axis=1)
pass_relation.head(20)
```

```python
pass_relation = pass_relation.groupby(['pos_min', 'pos_max']).sum().loc[:,'pass_count'].reset_index()
pass_relation
```

```python
pass_relation["team"] = pass_relation.apply(lambda x: eval(x["pos_min"])[0], axis=1)
pass_relation["player_min"] = pass_relation.apply(lambda x: eval(x["pos_min"])[1], axis=1)
pass_relation["player_max"] = pass_relation.apply(lambda x: eval(x["pos_max"])[1], axis=1)
pass_relation
```

```python
pass_relation["x"] = 0
pass_relation["y"] = 0
pass_relation["x_end"] = 0
pass_relation["y_end"] = 0
```

```python
for index, row in pass_relation.iterrows():
    pos_min = positions.loc[positions["team_player"] == (row["team"], row["player_min"]), :]
    pos_max = positions.loc[positions["team_player"] == (row["team"], row["player_max"]), :]
    
    pass_relation.at[index, 'x'] = pos_min.x.values[0]
    pass_relation.at[index, 'y'] = pos_min.y.values[0]
    pass_relation.at[index, 'x_end'] = pos_max.x.values[0]
    pass_relation.at[index, 'y_end'] = pos_max.y.values[0]
```

```python
pass_relation
```

```python
pass_relation.to_csv('../dashboard/src/stats/pass_relation.csv')
```
